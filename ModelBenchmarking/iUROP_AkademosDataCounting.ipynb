```python
import os
import matplotlib.pyplot as plt
q = "T"
def count_files_with_q(directory):
    count = 0
    for root, dirs, files in os.walk(directory):
        for file in files:
            if q in file:
                file_list.append(file[:9])
                count += 1
    return count

# Example usage
file_list = []

file_count = count_files_with_q("C:/Users/f-bon/Downloads/ImagesLabelling/All")
print(f"Number of files with 'T2' in the name: {file_count}")


print(file_list)

# Plotting the histogram
plt.figure(figsize=(12, 6))
plt.hist(file_list, bins=range(len(set(file_list)) + 1), edgecolor='black')
plt.title("Frequency of File Names with " +q)
plt.xlabel("File Name Prefix")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```
